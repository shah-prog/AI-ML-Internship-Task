```python
#pandas for handling data ,metabolic and seaborn for graph
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load iris dataset from seaborn
df = sns.load_dataset('iris')
```

```python
#it shape tell about the rows and column length
print("Shape:", df.shape)
```

```python
#dataset k columns batata ha
print(df.columns)
```

```python
#head pahli 5 rows day ga

print(df.head())
```

```python
#ya data type or missing values k baray main batata ha
df.info()
```

```python
#decribe statistic show krta ha mean medain etc
df.describe()
```

```python
#ya 2 features ka relation dikhati ha diff color main
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=df)
plt.title("Sepal Length vs Width")
plt.show()
```

```python
#Histogram data ki distribution show karta hai

df.hist(figsize=(10,8))
plt.show()
```

```python
#Boxplot se hum  (unusual values) identify kar sakte hain.
sns.boxplot(data=df)
plt.show()
```
